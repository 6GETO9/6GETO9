```python
# Загружаем изображение watermark

stampFile = '/content/ITHUB/челик.webp'
stamp = Image.open(stampFile)  # Загружаем изображение кота
stamp = stamp.convert('L')     # Приводим изображение к режиму черно-белого изображения 8 бит (0 ... 255)

stamp_width  = stamp.size[0]   # Получаем ширину
stamp_height = stamp.size[1]   # Получаем высоту

stamp_data = stamp.getdata()        # Получаем данные картинки в виде массива
stamp_data = np.array(stamp_data)   # Преобразуем в np массив

stamp = stamp_data.reshape(stamp_height, stamp_width)  # И решейпим массив с изображением в нужную конфигурацию
showImage(stamp)

print("Размерность:", stamp.shape)
print(stamp[100:101, 0:16])
```

```python
def pixelNormalize(p):
  p = round(p)
  if p > 255:
    p = 255
  if p < 0:
    p = 0
  return p
```

```python
#Накладываем watermark

out = image.copy()

width_space = 750  # Отступ слева
height_space = 1200 # Отступ снизу

t = 1 # Коэффициент прозрачности. Значение 2 соответствует 50%

# Находим координаты для начала наложения ватермарки
width_start  = width_space                                # Начало формированиа watermark слева
height_start = image_height - stamp_height - height_space # Начало формирования watermark снизу

# Пробегаемся по массиву и наносим ватермарку
for w in range(stamp_width):
  for h in range(stamp_height):
    pixel_r, pixel_g, pixel_b = image[height_start + h][width_start + w]  # Получим массив RGB для пикселя изображения кота
    pixel_stamp = float(stamp[h][w])  # Получим значение пикселя ватермарки

    r = pixelNormalize(pixel_r - (255 - pixel_stamp) / t)
    g = pixelNormalize(pixel_g - (255 - pixel_stamp) / t)
    b = pixelNormalize(pixel_b - (255 - pixel_stamp) / t)

    out[height_start + h][width_start + w] = [ r, g, b ]   # Положим значение пикселя обратно

showImage(out)
```

```python
# Записываем полученное изображение в файл
out = out.astype('uint8')
img_cat3 = Image.fromarray(out)
img_cat3.save("cat3-watermark.jpg", "JPEG")
```

```python
# Функция загрузки изображения
def loadImage(file_name, image_type='color'):
  image = Image.open(file_name) # Загружаем изображение из файла

  if image_type != 'color':
    image = image.convert('L')  # Приводим изображение к режиму черно-белого изображения 8 бит (0 ... 256)

  image_width  = image.size[0]  # Получаем ширину
  image_height = image.size[1]  # Получаем высоту

  image_data = image.getdata()          # Получаем данные картинки
  image_data = np.array(image_data)     # Преобразуем в np массив

  if image_type == 'color':
    image = image_data.reshape(image_height, image_width, 3)  # Решейпим изображение для цветного варианта
  else:
    image = image_data.reshape(image_height, image_width)     # Решейпим изображение для черно-белого варианта

  return image, image_width, image_height   # Возвращаем массив с пикселями и размерность изображениея


# Функция наложения изображения
# template - это изображение подложки (на которую будет наложено новое изображение)
# layer  - это изображение накладываемого слоя
# width  - оступ слева
# height - отступ снизу
def addLayer(template, layer, width_space, height_space):
  template_image  = template[0]
  template_width  = template[1]
  template_height = template[2]

  layer_image  = layer[0]
  layer_width  = layer[1]
  layer_height = layer[2]

  out_image = template_image.copy()

  # Находим координаты для начала наложения ватермарки
  width_start  = width_space                                   # Начало накладывания изображения слева
  height_start = template_height - layer_height - height_space # Начало накладывания изображения снизу

  # Пробегаемся по массиву и заменяем пиксели на подложке на пиксели из накладываемого слоя
  for w in range(layer_width):
    for h in range(layer_height):
      pixel = layer_image[h][w]  # Получим значение пикселя накладываемого изображения
      out_image[height_start + h][width_start + w] = [ pixel, pixel, pixel ]   # Заменим значение пикселя подложки на значение пикселя накладываемого слоя

  return out_image, template_width, template_height


# Функция формирования номера автомобиля
# num - номер автомобиля из букв и цифр
# region - регион автомобильного номера
def makeCarnumImage(num, region, debug = False):

  carnum = loadImage('/ITHUB/template.png');

  if debug:
    showImage(carnum[0])

  # ----------------------------- Номер машины -----------------------------------------

  # Определяем начальную точку, откуда будем формировать номер машины
  width_num  = 45
  height_num = 24

  # Количество пикселей между знаками в номере
  space = 15

  # Сформируем номер машины
  for s in num:
    s = s.lower() # Приводим символ к нижнему регистру

    symbol = loadImage('/ITHUB/9s.png/'+s+'.png', 'bw') # Загружаем черно-белое изображение

    carnum = addLayer(carnum, symbol, width_num, height_num) # Добавляем цифру на выходное изображение
    if debug:
      showImage(carnum[0]) # отобразим выходное изображения с очередной наложенной цифрой

    width_num  += symbol[1] + space

  # --------------------------- Регион номера машины---------------------------------------

  # Определяем начальную точку, откуда будем формировать номер машины
  width_num  = 438
  height_num = 54

  # Количество пикселей между знаками в номере
  space = 4

  # Если у нас только две цифры в номере региона, то заменяем отступы
  if len(str(region)) == 2:
    width_num += 21
    space += 2

  # Сформируем регион машины
  for s in region:
    symbol = loadImage('/ITHUB/7s.png/'+s+'s.png', 'bw') # Загружаем черно-белое изображение

    carnum = addLayer(carnum, symbol, width_num, height_num) # Добавляем цифру на выходное изображение
    if debug:
      showImage(carnum[0]) # отобразим выходное изображения с очередной наложенной цифрой

    width_num  += symbol[1] + space

  return carnum[0]
```
